```python
words = {}
import re
with open('cmudict.dict.txt', 'r') as f:
    for line in f.readlines():
        word, phonemes = line.strip().split(' ', 1)
        word = re.match(r'([^\(\)]*)(\(\d\))*', word).groups()[0]
        phonemes = phonemes.split(' ')
        syllables = sum([re.match(r'.*\d', p) is not None for p in phonemes])
        #print(word, phonemes, syllables)
        if word not in words:
            words[word] = []
        words[word].append({
            'phonemes': phonemes,
            'syllables': syllables
        })
```

```python
not_found = set()

import inflect
inflect_engine = inflect.engine()

def get_syllable_count(line):
    if not pd.isnull(line):
        counts = [0]
        line = line.lower()
        #print(line)
        ws = []
        for word in line.split(' '):
            '''if re.match(r'\w+', word):
                try:
                    x = inflect_engine.number_to_words(word).replace('-', ' ')
                except:
                    ws.append(word)
                    continue

                if (word == '0') or (x != 'zero'):
                    ws = ws + x.split(' ')
                else:
                    ws.append(word)
            else:
                ws.append(word)'''

            if re.search(r'\d', word):
                x = inflect_engine.number_to_words(word).replace('-', ' ')
                ws = ws + x.split(' ')
            else:
                ws.append(word)




        line = ' '.join(ws)
        #print(line)
        for word in line.split(' '):
            try:
                word = re.match(r'[\'"]*([\w\']*)[\'"]*(.*)', word).groups()[0]
                #print(word)
                if word:
                    if word not in words:
                        word = word.strip('\'')
                    syllables = set(p['syllables'] for p in words[word])
                    #print(syllables)
                    new_counts = []
                    for c in counts:
                        for s in syllables:
                            new_counts.append(c+s)

                    counts = new_counts
            except:
                not_found.add(word)
                #raise


        return set(counts)

#[ for w in "Memorial Day --"]

get_syllable_count("windless day 123rd -- 1 st david's 1st")
```
